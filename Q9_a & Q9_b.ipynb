```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline
import scipy
```

```python
df = pd.read_csv("Q9_a.csv")
df
```

```python
#for skewness
from scipy.stats import skew
scipy.stats.skew(df.speed, axis=0, bias=True)
```

```python
scipy.stats.skew(df.dist, axis=0, bias=True)
```

```python
# Importing library  
from scipy.stats import kurtosis
scipy.stats.kurtosis(df.speed, axis=0, fisher=True, bias=True)
```

```python
scipy.stats.kurtosis(df.dist,axis=0, fisher=True, bias=True)
```

```python
df.describe()
```

```python
plt.figure(figsize = (15,5))

plt.subplot(1,2,1)
plt.boxplot(df.speed)
plt.title('Speed')
plt.subplot(1,2,2)
plt.boxplot(df.dist)
plt.title('Dist')
plt.show()
```

```python
Inferences:
a)for speed 
1)the distribution is symmetric
2)There are no outliers
3)The data is concentrated around median.

b)for distance
1)the data is right skewed
2)there is one outlier 120
3)The data is concentarted around the median.
```
